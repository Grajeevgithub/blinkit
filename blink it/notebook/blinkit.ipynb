```python
import pandas as pd 
import numpy as np
import seaborn as sns 
import matplotlib.pyplot as plt
```

# trends byy linechart

```python
data["Sales (Lakhs)"] = data["Sales"] / 100000
```

```python


# Grouping for trend
trend_data = data.groupby("Outlet Establishment Year")["Sales"].mean().reset_index()

# Plotting
plt.figure(figsize=(10, 6))
sns.lineplot(data=trend_data, x="Outlet Establishment Year", y="Sales", marker="o", color="green")
plt.title("Average Sales Trend by Outlet Establishment Year", fontsize=14)
plt.xlabel("Outlet Establishment Year")
plt.ylabel("Average Sales")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# boxplot

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.boxplot(data=data, x="Outlet Type", y="Sales", palette="Set2")
plt.title("Sales Distribution by Outlet Type")
plt.xlabel("Outlet Type")
plt.ylabel("Sales")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# scatterplot

```python
data["Sales (K)"] = data["Sales"] / 100000
```

```python


plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x="Item Visibility", y="Sales", hue="Item Type", alpha=0.7)
plt.title("Sales vs Item Visibility")
plt.xlabel("Item Visibility")
plt.ylabel("Sales")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
sns.scatterplot(data=data, x="Item Weight", y="Sales")
plt.title("Sales vs Item Weight")
plt.show()
```
